Let's compute some spectrograms as examples.

Let's begin with the well-known chirp

```python
from scipy import signal
#let's begin creating a linear chirp

fs = 1000 #1KHz sampling frequency
ts = 1./fs
t = arange(0,10,ts)

y_linear = 2*signal.chirp(t,3,10,30)
y_log = 2*signal.chirp(t,3,10,30,method = 'quadratic')

plot(t,y_linear)
figure()
plot(t,y_log)
```

Now let's pretend that the signal is indeed stationary and compute the periodogram

```python
f, Pxx_linear = signal.periodogram(y_linear, fs)
f, Pxx_log = signal.periodogram(y_log,fs)

#let's plot

idx_f = f < 30
plot(f[idx_f],Pxx_linear[idx_f])
xlabel('f [Hz]')
ylabel('Periodogram')
title('Linear')
figure()
plot(f[idx_f],Pxx_log[idx_f])
xlabel('f [Hz]')
ylabel('Periodogram')
title('Quadratic')
```
