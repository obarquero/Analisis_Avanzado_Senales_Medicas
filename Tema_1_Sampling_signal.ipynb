If a function $f(t)$ contains no frequencies higher than $F_n$ Hz, it is completely determined by giving its ordinates at a series of points spaced $1/(2F_n)$ seconds apart.

So this statement allows us to sampling a signal $x(t)$ at a samplig frequency $F_s$ higher than $2F_N$; $F_s > 2F_N$. Where $F_N$ is the maximum frequency contained in the signal, and called Nyquist frequency.

A signal $x(t)$ can be sampled resulting in $x(n) = x(nT_s)$, where $T_s = 1/F_s$

```python
#Let's play a little bit with the sampling theorem.

import numpy as np

#let's create a signal with a very high f_s, as if it were continuous.

f_sc = 1e4 #"continuous" sampling frequency
t_sc = 1./f_sc

t = np.arange(0,5,t_sc) #in seconds

#two sinusoides
f1 = 3 #Hz F_n
f2 = 2.4 #Hz
y1 = np.sin(2*np.pi*f1*t)
y2 = 0.5*np.sin(2*np.pi*f2*t + 0.54)

y = y1+y2

plot(t,y)
```

First, let's look at the fourier domain

```python
#Compute the fft of y

y_fft = np.fft.fft(y)

#freq axis
freq = np.fft.fftfreq(t.shape[-1],t[1]-t[0])

#plot
plot(freq, abs(y_fft))
```

```python
#lets zoom a little bit
idx = [abs(freq) < 5]

figure
plot(freq[idx],abs(y_fft[idx]))
```

Now we are going to sample and subsample the original signal, and then try to reconstruc it from the sampled version

```python
#Let's sampled the signal using the double the Nyquist frequency

fn = f1
fs = 10 #sampling frequency
#print fs
T = 1./fs
#get the number of samples in the older f_sc
n_sampled = round(T*f_sc)
#print(n_sampled)
y_sampled = y[0:-1:n_sampled]
t_sampled = t[0:-1:n_sampled]
#print len(y_sampled)

plot(t_sampled,y_sampled,t,y)
```

```python
#let's plot only the stems

stem(t_sampled,y_sampled)
plot(t,y,color = 'k',linestyle = ':',linewidth  = 3)
```

Now let's subsampling the original signal. Let's say we pick only samples each 2 seconds

```python
fn = f1
fsubs = 2 #sampling frequency
#print fs
Tsubs = 1./fsubs
#get the number of samples in the older f_sc
n_subsampled = round(Tsubs*f_sc)
#print(n_sampled)
y_subsampled = y[0:-1:n_subsampled]
t_subsampled = t[0:-1:n_subsampled]
#print len(y_sampled)

plot(t_subsampled,y_subsampled,t,y)
```

```python
#let's plot only the stems

stem(t_subsampled,y_subsampled,color = 'magenta')
plot(t,y,color = 'k',linestyle = ':',linewidth  = 3)
```

Now let's compare the reconstruction using sincs, that is using the Whittake-Shannon interpolation formula

```python
def sinc_interp(x, s, u):
    """
    Interpolates x, sampled at "s" instants
    Output y is sampled at "u" instants ("u" for "upsampled")
    
    from Matlab:
    http://phaseportrait.blogspot.com/2008/06/sinc-interpolation-in-matlab.html        
    """
    
    if len(x) != len(s):
        raise Exception, 'x and s must be the same length'
    
    # Find the period    
    T = s[1] - s[0]
    
    sincM = tile(u, (len(s), 1)) - tile(s[:, newaxis], (1, len(u)))
    y = dot(x, sinc(sincM/T))
    return y

y_sinc_in = sinc_interp(y_sampled,t_sampled,t)

plot(t,y_sinc_in,t,y)
plot(t_sampled,y_sampled,'o')
```

```python
y_sinc_int_sub = sinc_interp(y_subsampled,t_subsampled,t)
plot(t,y_sinc_int_sub,t,y)
plot(t_subsampled,y_subsampled,'o')
```
