Consider the following random process: $x(n) = A\sin(n\omega_0 + \phi)$

Where: $A$ and $\omega_0$ are fixed constants and $\phi$ is a random variable uniformly distributed over the interval $-\pi$ and $\pi$. That is the pdf of $\phi$ is:
\begin{equation}
f_\phi(\alpha) = \begin{cases} (2\pi)^{-1}, & \mbox{if } -\pi\le\alpha<\pi \\ 0, & \mbox{otherwise}\end{cases}
\end{equation}

```python
#Let's represent some realizations

import numpy as np
import matplotlib.pyplot as plt

#define the constants
A = 2
w_0 = 2*np.pi*50
n = np.linspace(0,20,100)
num_real = 10000
#define the r.v.
np.random.seed(5)
phi = np.random.uniform(-np.pi,np.pi,num_real)

#define the process
x = np.zeros((100,num_real))
for m in range(len(phi)):
    x[:,m] = A*np.sin(n*w_0 + phi[m])

#plotting
offset = 0
for m in range(5):
    plt.plot(n,x[:,m]+offset)
    offset = offset - 5
```

Let's compute some statistics:

Mean
\begin{equation}
\mathbb{E}\left[x(n)\right] = \int^{\infty}_{-\infty} Asin(n\omega_0 + \alpha)\ f_\phi(\alpha)\ d\alpha = \\
 = \int^{\pi}_{-\pi} \frac{1}{2\pi}\ Asin(n\omega_0 + \alpha)\ d\alpha = 0 
\end{equation}

```python
#Let's compute the mean of the process
#We are going to use the sample mean as a estimator of the expected value

#Let's compute the mean value for n = 0 and n = 50
mu_0 = np.mean(x[0,:])
mu_50 = np.mean(x[50,:])

print("Mean value for n = 0: mu_0 = " + str(mu_0))
print("Mean value for n = 0: mu_0 = " + str(mu_50))
```
