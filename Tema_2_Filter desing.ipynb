Let's make some plots of the zero-pole plane.

We are going to use the followin transfer function:

\begin{equation}
H(z) = \frac{1}{2}(1-e^{j\pi/3}z^{-1})(1-e^{-j\pi/3}z^{-1}) = \frac{1}{2}-\frac{1}{2}z^{-1}+\frac{1}{2}z^{-2}
\end{equation}

\begin{equation}
