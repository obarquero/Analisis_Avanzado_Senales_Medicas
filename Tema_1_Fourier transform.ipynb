```python
#sinusoidal signal
import numpy as np

#let's create a signal with a very high f_s, as if it were continuous.

f_sc = 1e4 #"continuous" sampling frequency
t_sc = 1./f_sc

t = np.arange(0,5,t_sc) #in seconds

#two sinusoides
f1 = 3 #Hz F_n
f2 = 2.4 #Hz
y1 = np.sin(2*np.pi*f1*t)
y2 = 0.5*np.sin(2*np.pi*f2*t + 0.54)

y = y1+y2

plot(t,y)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
#Compute the fft of y

y_fft = np.fft.fft(y)

#freq axis
freq = np.fft.fftfreq(t.shape[-1],t[1]-t[0])

#plot

#lets zoom a little bit
idx = [abs(freq) < 5]

figure
plot(freq[idx],abs(y_fft[idx]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('|FFT|')
```

Let's compute a periodogram of a noisy sinusoid. Generate a test signal, a 2 Vrms sine wave at 1234 Hz, corrupted by 0.001 $V^2/Hz$ of white noise sampled at 10 kHz.

```python
from scipy import signal

fs = 10e3
N = 1e5
amp = 2*np.sqrt(2)
freq = 100.0
noise_power = 0.001 * fs / 2
time = np.arange(N) / fs
x = amp*np.sin(2*np.pi*freq*time)
x += np.random.normal(scale=np.sqrt(noise_power), size=time.shape)

plot(time,x)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
#let's zoom a little bit
idx = time < 0.125
figure()
plot(time[idx],x[idx])
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

Compute and plot the power spectral density.

```python
f, Pxx_den = signal.periodogram(x, fs)
semilogy(f, Pxx_den)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
axis('tight')
ylim([1e-8, 1e2])
#let's zoom a little bit
idx = f < 150
figure()
semilogy(f[idx],Pxx_den[idx])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
axis('tight')
ylim([1e-8,1e2])
```
