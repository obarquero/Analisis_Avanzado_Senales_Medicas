## Loading data in python

First we load the ecg. Here we can have several different formats in which the ECG samples are stored. In this simple case the ECG is stored as txt file, so simply call numpy method to read txt files.

```python
import numpy as np
import matplotlib.pyplot as plt

#Load ecg sample
file_path = './files/data/pat1.dat'

#read txt files
ecg1 = np.loadtxt(file_path)
fs = 300.
#ploting the ecg
plt.plot(ecg1)
plt.axis('tight')
```

Some things to note in the ecg:

    1) There is no xlabel, and ylabel
    2) We cannot see anything, so let's zoom in

```python
plt.clf()
figure(figsize=(30,8))
plt.plot(ecg1[:40000])
plt.figure()
plt.plot(ecg1[10001:30000])
plt.figure()
plt.plot(ecg1[350:800])
```

Let's zoom in y axis to see the 50 Hz noise

```python
plt.plot(ecg1[350:800])
a = [0,450,-.6,1]
plt.axis(a)
```

```python
#Lets compute the fft of the beat

beat = ecg1[350:800]
t_beat = np.arange(0,len(beat))
t_beat = t_beat/fs
fft_beat  = np.fft.fft(beat-mean(beat))

#freq axis
freq = np.fft.fftfreq(t_beat.shape[-1],t_beat[1]-t_beat[0])
idx = logical_and(freq >= 0,freq <= 60)
plot(freq[idx],abs(fft_beat[idx]))
```

```python
#create some low-pass filter
from scipy import signal
fn = fs/2.
wc = 30/fn
b,a = signal.butter(4,wc,'low')

#filter the signal
ecg_filtered = signal.filtfilt(b,a,ecg1)

#plotting
plot(ecg1[350:800])         
plot(ecg_filtered[350:800],color = 'green')
a = [0,450,-.5,.1]
axis(a)


#plotting the spectrum 
fft_beat_filtered  = np.fft.fft(ecg_filtered[350:800]-mean(ecg_filtered[350:800]))
figure()
plot(freq[idx],abs(fft_beat[idx]))
plot(freq[idx],abs(fft_beat_filtered[idx]),color = 'black')
```
